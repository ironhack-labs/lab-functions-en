## 9. Write a function that calculates the standard deviation of a list.
`NOTE: Do not use any libraries or already built functions. üòâ`

```python
def st_dev(list_sd):
#your code here
    n = len(list_sd)  
    mean = sum(list_sd) / n  
    variance = sum((x - mean) ** 2 for x in list_sd) / (n - 1)  
    return variance ** 0.5
```

```python
# This will test your function 
test_stdev(st_dev)
```

## 10. Write a function to check if a string is a pangram, i.e.: if it contains all the letters of the alphabet at least once. Mind that the strings may contain characters that are not letters.

```python
from string import ascii_lowercase
def pangram(string):
#your code here
    return set(ascii_lowercase).issubset(set(string.lower()))
```

## 11. Write a function that receives a string of comma separated words and returns a string of comma separated words sorted alphabetically.

`NOTE: You may use sorted but not split and definitely no join! ü§™`

```python
def sort_alpha(string):
#your code here
    words = [word for word in string] if "," not in string else eval(f"[{string}]")  
    return ",".join(sorted(words)) 
```

```python
# This will test your function 
test_alpha(sort_alpha)
```

## 12. Write a function to check if a given password is strong (at least 8 characters, at least one lower case, at least one upper case, at least one number and at least one special character). It should output True if strong and False if not.

```python
import re
def check_pass(password):
#your code here
    return (
        len(password) >= 8 and
        re.search(r"[a-z]", password) and  # At least one lowercase letter
        re.search(r"[A-Z]", password) and  # At least one uppercase letter
        re.search(r"\d", password) and     # At least one digit
        re.search(r"[#@!$%&()^*[\]{}]", password)  # At least one special character
    ) is not None  # Ensures it returns True/False
```

```python
# This will test your function 
test_pass(check_pass)
```
